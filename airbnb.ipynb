```python
import pandas as pd
import json

import pymongo


# Function to extract availability values from the string
def extract_availability(row):
    try:
        availability_dict = json.loads(row.replace("'", "\""))  # Replace single quotes with double quotes for valid JSON
        return availability_dict
    except:
        return {
            'availability_30': 0,
            'availability_60': 0,
            'availability_90': 0,
            'availability_365': 0
        }

# Apply the extract_availability function to the entire 'availability' column
df['availability'] = df['availability'].apply(extract_availability)

# Access availability values for all rows
df['availability_30'] = df['availability'].apply(lambda x: x['availability_30'])
df['availability_60'] = df['availability'].apply(lambda x: x['availability_60'])
df['availability_90'] = df['availability'].apply(lambda x: x['availability_90'])
df['availability_365'] = df['availability'].apply(lambda x: x['availability_365'])
```

```python
# Function to parse the address string to a dictionary
def parse_address(address_str):
    try:
        if isinstance(address_str, str):
            return json.loads(address_str.replace('$numberDouble', '"numberDouble"').replace('false', 'false').replace('true', 'true'))
        elif isinstance(address_str, dict):
            return address_str
        else:
            return {}
    except json.JSONDecodeError:
        return {}  # Return an empty dictionary if parsing fails

# Apply the parsing function to each element in the Series
parsed_addresses = df['address'].apply(parse_address)

# Create new columns in the DataFrame for each extracted field
df['street'] = parsed_addresses.apply(lambda x: x.get('street', ''))
df['suburb'] = parsed_addresses.apply(lambda x: x.get('suburb', ''))
df['government_area'] = parsed_addresses.apply(lambda x: x.get('government_area', ''))
df['market'] = parsed_addresses.apply(lambda x: x.get('market', ''))
df['country'] = parsed_addresses.apply(lambda x: x.get('country', ''))
df['country_code'] = parsed_addresses.apply(lambda x: x.get('country_code', ''))

# Accessing coordinates within the location dictionary if it exists
df['latitude'] = parsed_addresses.apply(lambda x: x.get('location', {}).get('coordinates', [])[0] if 'location' in x and 'coordinates' in x['location'] else None)
df['longitude'] = parsed_addresses.apply(lambda x: x.get('location', {}).get('coordinates', [])[1] if 'location' in x and 'coordinates' in x['location'] else None)
```
