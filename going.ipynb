```python
import pandas as pd
import json

# Load the CSV file into a DataFrame
df = pd.read_csv("C:\\Users\\user\\Desktop\\airbnb\\abnb_dataset.csv")

# Function to extract availability values from JSON string
def extract_availabilities(avail_str):
    try:
        availability_dict = json.loads(avail_str.replace("'", "\""))
        availability_30 = availability_dict.get('availability_30', None)
        availability_60 = availability_dict.get('availability_60', None)
        availability_90 = availability_dict.get('availability_90', None)
        availability_365 = availability_dict.get('availability_365', None)
        return availability_30, availability_60, availability_90, availability_365
    except Exception as e:
        print(f"Error extracting availabilities: {e}")
        return None, None, None, None

# Apply the function to the 'availability' column and create new columns for each availability
df['availability_30'], df['availability_60'], df['availability_90'], df['availability_365'] = zip(*df['availability'].map(extract_availabilities))

# Print the DataFrame with the extracted availabilities
print(df[['availability_30', 'availability_60', 'availability_90', 'availability_365']])
```

```python
import pandas as pd
import json

# Read the CSV file
df = pd.read_csv("C:\\Users\\user\\Desktop\\airbnb\\abnb_dataset.csv")

# Function to safely convert string to dictionary
def parse_address(address):
    try:
        return json.loads(address.replace("'", '"'))
    except (json.JSONDecodeError, AttributeError):
        return {}

# Convert 'address' column from string to dictionary
df['address'] = df['address'].apply(parse_address)

# Access 'suburb' and 'government_area' values
suburb_values = df['address'].apply(lambda x: x.get('suburb', None))
government_area_values = df['address'].apply(lambda x: x.get('government_area', None))

print(suburb_values)
print(government_area_values)
```

```python
import pandas as pd
import re

# Read the CSV file
df = pd.read_csv("C:\\Users\\user\\Desktop\\airbnb\\abnb_dataset.csv")

# Define a function to extract 'suburb' and 'government_area' values using regular expressions
def extract_suburb_and_government_area(address_str):
    try:
        # Define the pattern to match the suburb and government_area values
        pattern = r'"suburb":\s*"([^"]+)",\s*"government_area":\s*"([^"]+)"'
        match = re.search(pattern, address_str)
        if match:
            return match.group(1), match.group(2)
        else:
            return None, None
    except AttributeError:
        return None, None

# Apply the function to extract 'suburb' and 'government_area' values
df['suburb'], df['government_area'] = zip(*df['address'].apply(extract_suburb_and_government_area))

# Display the values
print(df['suburb'])
print(df['government_area'])
```

```python
import pandas as pd
import ast

# Read the CSV file
df = pd.read_csv("C:\\Users\\user\\Desktop\\airbnb\\abnb_dataset.csv")

# Define a function to extract 'suburb' and 'government_area' values
def extract_suburb_and_government_area(address_str):
    try:
        address_dict = ast.literal_eval(address_str)
        suburb = address_dict.get('suburb', None)
        government_area = address_dict.get('government_area', None)
        return suburb, government_area
    except (ValueError, SyntaxError):
        return None, None

# Apply the function to extract 'suburb' and 'government_area' values
df['suburb'], df['government_area'] = zip(*df['address'].apply(extract_suburb_and_government_area))

# Display the values
print(df['suburb'])
print(df['government_area'])
```
